```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
xs = [10 ** (i) for i in range(-3, 2)] 
ys = [2 * x ** (-0.074) for x in xs] 
```

```python
plt.xscale('log')
plt.yscale('log')

plt.plot(xs, ys)
plt.gca().set_yticks(np.arange(max(ys), min(ys), 0.2))

plt.gca().set_yticklabels(np.arange(max(ys), min(ys), 0.2))
plt.savefig('reproduced_scaling_line.png')
```
